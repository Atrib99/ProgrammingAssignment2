```python
import yfinance as yf
import pandas as pd
import mplfinance as mpf
import matplotlib.pyplot as plt

start_date = "2023-01-01"
end_date = "2024-01-01"
symbol = "btc-usd"
# interval = "1mo"

df = yf.download(symbol, start_date, end_date, interval = "1d")
print(df)
# Plot the closing prices and EMA
mpf.plot(df, type='candle', mav=(200,), volume=False, title='Trading Strategy', ylabel='Price')
        #  addplot=[mpf.make_addplot(df['EMA21'], color='orange', width=1.2)])
plt.show()
ema = 100
count = 0
df['EMA21'] = df['Close'].ewm(span=ema, adjust=False).mean()
for i in range(2,len(df)):
   if (df['Close'][i] >= df['EMA21'][i] and
       df['Close'][i-1] > df['EMA21'][i-1] and
       df['Close'][i-2] > df['EMA21'][i-2] and
       df['Low'][i] <= df['EMA21'][i-1] ):
    count += 1
print("number of trades executed = ",count)
```
