```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a sample dataset (random normal distribution)
np.random.seed(42)
mean = 50
std_dev = 10
sample_size = 1000
sample_data = np.random.normal(mean, std_dev, sample_size)
```

```python
# Define the number of simulations
num_simulations = 1000
```

```python
# Initialize an array to store simulation results
simulation_results = []
```

```python
# Perform Monte Carlo simulation
for _ in range(num_simulations):
    # Randomly sample from the dataset with replacement
    sampled_data = np.random.choice(sample_data, size=sample_size, replace=True)

    # Calculate the mean of the sampled data
    simulation_results.append(np.mean(sampled_data))
```

```python
 # Plot the histogram of simulation results
plt.hist(simulation_results, bins=30, density=True, alpha=0.7, color='blue', label='Monte Carlo Simulation')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='True Mean')
plt.title('Monte Carlo Simulation')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.show()
```
